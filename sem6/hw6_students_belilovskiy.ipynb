## Задача 1.

Пусть $X_1,...,X_n$ выборка из некоторого распределения $\mathrm{P}$, причем ${\sf D} X_1=\sigma^2<+\infty$, и $\sigma$ неизвестно.
Рассмотрим статистику $S^2=\frac1n\sum\limits_{i=1}^n(X_i-\overline{X})^2$.

1. Докажите, что статистика $S^2$ равна $\overline{X^2}-\overline{X}^2$.

По определению среднего: $\overline{X} = \frac 1n \sum\limits_{i=1}^nX_i$, откуда подставив выражение выше можем получить:
$$
S^2 = \frac 1n \sum\limits_{i=1}^n(X_i^2-2X_i\overline{X}+\overline{X}^2) =
\overline{X^2} - 2\overline{X}^2 + \overline{X}^2 = \overline{X^2} - \overline{X}^2
$$

2. Является ли она состоятельной оценкой $\sigma^2$? Для ответа на этот вопрос воспользуйтесь утверждениями с лекции.

Да, она является состоятельной оценкой для дисперсии σ^2, это показывает, что с увеличением размера выборки оценка $S^2$ сходится к истинному значению дисперсии $\sigma^2$ по вероятности. Данное свойство проиллюстрировано на графиках в задаче 6. 

Также из графиков видно свойство асимптотической нормальности, из чего следует *состоятельность* оценки $S^2$

3. Является ли статистика $S^2$ несмещенной оценкой $\sigma^2$? Для ответа на этот вопрос вспомните различные варианты определения дисперсии.

Как известно из курса теории вероятностей, существуют эквивалентные определения дисперсии:
$$
\mathbb D X = \mathbb E \left( X - \mathbb E X\right)
$$
Откуда в силу *линейности* математического ожидания $\mathbb E$ случайной велечины $X$ следует соотношение:
$$
\mathbb D X = \mathbb E X^2 - (\mathbb E X)^2
$$
Оценка $\widehat{\theta}$ называется *несмещённой*, если $\forall \theta \in \Theta\colon \mathsf{E}_\theta \widehat{\theta} = \theta$. 

Пусть $\mu$ и $\overline{X}$ -- среднее и его оценка соответственно. Прибавим и вычтем его, чтобы затем разбить на две части:
$$
S^2=\frac1n\sum\limits_{i=1}^n(X_i-\overline{X})^2 =\frac1n\sum\limits_{i=1}^n((X_i - \mu) -(\overline{X} - \mu))^2 = \frac1n\sum\limits_{i=1}^n(X_i-\mu)^2 - 2(\overline{X} - \mu)\frac1n\sum\limits_{i=1}^n(X_i-\mu)^2 + \frac nn (\overline{X} - \mu)^2
$$

Заметим, что: $\frac1n\sum\limits_{i=1}^n(X_i-\mu) = \overline{X} - \mu$ Тогда преобразуем выражение выше:
$$
S^2 = \frac1n\sum\limits_{i=1}^n(X_i-\mu)^2\ -\ (\overline{X} - \mu)^2
$$
Теперь перейдём к матожиданию и воспользуемся его линейностью: $\mathbb E S^2 = \mathbb E \left[\frac1n\sum\limits_{i=1}^n(X_i-\mu)^2 \right] + \mathbb E \left[(\overline{X} - \mu)^2 \right]$

Первое слагаемое по определению является истинной дисперсией $\sigma^2$, разберёмся со вторым слагаемым:

$$
\mathbb E \left[ \left(\overline{X} - \mu \right)^2 \right] = \mathbb E \left[\left(\frac 1n \sum\limits_{i=1}^nX_i - \mu \right)^2 \right] = \mathbb E \left[ \frac{1}{n^2} \left(\sum\limits_{i=1}^nX_i - \mu \right)^2 \right] = \mathbb E \left[ \frac{1}{n^2} \left(\sum\limits_{i=1}^n(X_i - \mu)^2 + 2\sum\limits_{i=1, j=1, i<j}^{n}(X_i - \mu)(X_j - \mu) \right) \right]
$$
Для независимых случайных величин $X_i , X_j$ вторая сумма будет равна нулю. В итоге получаем:
$$
\mathbb E \left[ \left(\overline{X} - \mu \right)^2 \right] = \mathbb E \left[ \frac{1}{n^2} \left(\sum\limits_{i=1}^n(X_i - \mu)^2 \right) \right] = \frac1n \mathbb E \left[ \frac1n \left(\sum\limits_{i=1}^n(X_i - \mu)^2 \right) \right] = \frac 1n \sigma^2
$$
Соберём наш результат:
$$
\mathbb E S^2 = \frac{n-1}{n} \sigma^2 \neq \sigma^2
$$
Таким образом полученная является *смещённой* оценкой $\sigma^2$. При этом легко заметить, что при $n \to \infty$ $\mathbb E S^2 \to \sigma^2$, поэтому *асимптотически несмещённой* данная оценка уже является

По Усиленному закону больших чисел (УЗБЧ) данная оценка является *сильно состоятельной*, то есть выполнена сходимость *почти наверное* $\forall \theta \in \Theta\colon \mathsf{E}_\theta \widehat{\theta} \to \theta$. Отсюда следует также её обычная состоятельность и несмещённость (свойства сходимостей, *почти наверное* -- самая сильная) и выполнено $\mathbb E \widehat{\theta} = \theta$. Рассмотрим модель линейной регрессии

где $Y \in \mathbb{R}^n$ - отклик, $X \in \mathbb{R}^{n \times d}$, $\theta \in \mathbb{R}^d$, $ɛ \in \mathbb{R}^n$ - шум, $n > d$, $rk X = d$.

  $$\widehat{\theta} = (X^TX)^{-1}X^T Y$$

  является несмещенной оценкой параметра $\theta$. Будет ли выполнено свойство несмещенности для Ridge-регрессии?

  *Указание:* используйте свойство линейности математического ожидания в матричном виде.

3. Покажите, что, если дополнительно матрица ковариаций $D ɛ = σ^2 I_n$, то дисперсия МНК-оценки: $$D \widehat{\theta} = \sigma^2 (X^TX)^{-1}$$ Что показывает данный результат в случае мультиколлинеарности признаков? 

  *Указание:* используйте свойство дисперсии в матричном виде.

**Напоминание**

1. Матричное свойство математического ожидания:

  $∀ A \in \mathbb{R}^{m \times k}$, случайного вектора $\psi \in \mathbb{R}^k$

  $$E A \psi = A E \psi$$
2. Матричное свойство дисперсии:
  $∀ A \in \mathbb{R}^{m \times k}$, случайного вектора $\psi \in \mathbb{R}^k$

  $$D A \psi = A (D \psi) A^T$$


**Пункт 1**

Случайными велечинами здесь являются векторы $Y$ и $\varepsilon$, $\varepsilon \sim \mathcal N(o, \sigma^2)$. Неизвестной величиной является вектор $\theta$

**Пункт 2**

Возьмём матожидание от выражения $Y = X \theta + \varepsilon$ и воспользуемся его линейностью, учитывая, что $\mathbb E \varepsilon = 0$, что следует из $\varepsilon \sim \mathcal N(o, \sigma^2)$

Тогда $\mathbb E Y = X \theta $, домножив правую и левую часть слева на $X^T$ получаем выражение: $X^T \mathbb E Y = X^TX \theta $

Матожидание МНК-оценки: $\widehat{\theta} = (X^TX)^{-1}X^T Y$:
$$
\mathbb E \widehat{\theta} = \mathbb E (X^TX)^{-1}X^T Y = (X^TX)^{-1}X^T \mathbb E Y
$$
Подставив полученный выше результат, прелбразуем наше выражение:
$$
\mathbb E \widehat{\theta} = (X^TX)^{-1}X^T(X^T)^{-1} X^TX \theta = \theta
$$
То есть МНК - оценка $\theta$ действительно является **несмещённой**

Для Ridge-регрессии свойство несмещённости выполенно не будет. Аналогично предыдущим рассужениям используя полученный ранее результат Ridge-оценки: $\widehat{\theta} = (X^TX + \lambda E)^{-1}X^T Y$ Приходим к выражению:
$$
\mathbb E \widehat{\theta} = (X^TX + \lambda E)^{-1}X^T(X^T)^{-1} X^TX \theta \neq \theta
$$ где $E$ -- Единичная матрица размера $dxd$

Таким образом, оценка $\theta$, полученная из Ridge-регрессии является **смещённой**

**Пункт 3**

Если дополнительно задана матрица ковариаций $D ɛ = σ^2 I_n$, то Дисперсия окажется ненулевой. Причём случайность определяется величиной $\varepsilon \Rightarrow \mathbb D Y = \mathbb D \varepsilon$. Для поиска дисперсии воспользуемся его матричным свойством $D A \psi = A (D \psi) A^T$. Тогда:
$$
\mathbb D \widehat{\theta} = \mathbb D \left[(X^TX)^{-1}X^TY \right] = (X^TX)^{-1}X^T \mathbb D \left[ Y \right] X(X^TX)^{-1} = σ^2(X^TX)^{-1}X^T X(X^TX)^{-1} = σ^2(X^TX)^{-1}
$$

В случае мультиколлинеарности признаков определитель матрицы регрессоров $X$, а заначит и мтарицы $X^TX$ будет близок к нулю. В выражении для дисперсии находится матрица обратная к данной, а значит для полученной оценки будет справедливо:
$$
\mathbb D \widehat{\theta} >> \theta
$$
Таким образом в случае мультиколлинеарности оценка не может быть вычислена с помощью обычной линейной регрессии.

## Задача 5


Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $\mathcal{N}(0, \sigma^2)$. Известно, что в качестве оценки параметра $\sigma^2$ можно использовать следующие оценки $S^2, \frac{n}{n-1}S^2$. Проверьте эти оценки на несмещенность повторив эксперимент с семинара.

Реализуйте две функции, каждая из которых на вход берет несколько выборок, а на выход выдает массив оценок  
(первая функция - $S^2$, вторая функция - $\frac{n}{n-1}S^2$)  для каждой выборки.  
*Примечание:* У матрицы, заданной в виде `numpy.array` есть метод `var` вычисляющий выборочную дисперсию.

```python
def estimate_s2(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    S=[]
    for m in range(len(X[:,1])):
        S.append(np.var(X[m]))
    return S

def estimate_s2_corrected(X):
    '''
    Принимает на вход массив размером (n_samples, sample_size), выдает массив оценок размера (n_samples,)
    '''
    S=[]
    for m in range(len(X[:,1])):
        S.append(np.var(X[m])*(len(X[m])/(len(X[m])-1)))
        
    return S
```

Проверьте правильность своей реализации

```python
X = np.array([[1, 3, 3, 6], [2, 5, 1, 89]])
estimate_s2(X)

assert(sum(estimate_s2(X) != np.array([3.1875, 1399.6875])) == 0)
assert(sum(estimate_s2_corrected(X) != np.array((4.25, 1866.25))) == 0)
```

Задайте список оценок и разные параметры для отрисовки графика

```python
estimators = [
    (estimate_s2, r'$S^2$', 'purple', 0.1),
    (estimate_s2_corrected, r'$\frac{n}{n-1}S^2$', 'green', 0.2)
]
```

Создайте 500 выборок размера `sample_size`.

```python
sample_count = 500
sample_size = 100
X = sps.norm.rvs(size = (sample_count, sample_size))
X.shape
```

Визуализируйте полученные оценки на одном графике. Средние значения по 500 экспериментам для каждой оценки нужно обозначить звездочками.

```python
def est_plot(distribution, estimators, sample_size, sample_count):
    '''
    Построение графика разброса реализаций оценок и их средних значений.

    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    sample_size -- размер выборок
    sample_count -- количество генерируемых выборок
    '''
    X = distribution.rvs(size=(sample_count, sample_size))
    plt.figure(figsize=(15, 1*len(estimators)))
    for estimator, name, color, y in estimators:
        E = np.array(estimator(X))
        plt.scatter(E, np.zeros(sample_count) + y, alpha=0.1,
                        s=100, color=color, label=name)
        plt.scatter(E.mean(), y, marker='*', s=300,
                        color='w', edgecolors='black')

        plt.vlines(1, -1, 1, color='r')
        plt.title('Размер выборки = %d' % sample_size)
        plt.yticks([])
        plt.legend(loc='lower right')
        plt.xlim((0.8, 1.2))
        plt.ylim((-0.1, 0.2 * len(estimators)))
```

Изучите поведение среднего оценок из первого пункта при росте размера $n$ выборки - постройте график зависимости среднего оценки по всем экспериментам от размера выборки. Для вычисления зависимости нужно один раз сгенерировать выборки из достаточно большого размера и посчитать оценки по префиксам, используя функции из `numpy`.




```python
def mean_plot(distribution, estimators, n_grid, sample_count):
    '''
    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    n_grid -- массив значений размера выборки
    sample_count -- количество генерируемых выборок
    '''
    Est=np.zeros((len(n_grid),2)) #Массив нулей размера: Длина_Списка_размеров_выборок x Число функций оценки
    #Он нужен для записи данных оценки
    plt.figure(figsize=(15, 2*len(estimators)))
    for j in range(len(n_grid)):
        X = distribution.rvs(size=(sample_count, int(n_grid[j])))
        for i in range(len(estimators)):
            E = np.array(estimators[i][0](X))
            Est[j][i] = E.mean() #Записываем значение средней оценки для каждой функции-оценки
    
    #Строим график и делаем подписи
    plt.scatter(n_grid,Est[:, 0], color='#ED5300', label=estimators[0][1], s=60, alpha=0.7) 
    plt.scatter(n_grid,Est[:, 1], color ='#6A009F', label=estimators[1][1], s=60, alpha=0.7)
    plt.xlabel('Размер выборки $n$')
    plt.ylabel('Средняя оценка')
    plt.legend(loc='lower right')
```

```python
n_grid = np.linspace(10, 500, 200) #Массив размеров выборок для анализа
mean_plot(sps.norm, estimators, n_grid, sample_count)
```

```python
sample_size2 = 300  # размер выборок
sample_count2 = 500  # количество выборок

samples2 = sps.norm.rvs(size=(sample_count2, sample_size2))
samples2.shape
```

```python
estimation = np.array([np.var(samples2[:, :i], axis=1) for i in range(1, sample_size2+1)]).T
estimation.shape
```

```python
def est_plot(estimation, sample_count, sample_size, left=None, right=None, xlim = sample_size2 ):
    '''
    estimation -- массив оценок от размера выборки
    sample_count -- количество генерируемых выборок
    sample_size -- размер каждой выборки
    left, rigth -- границы доверительного интеравала, нужны будут далее
    xlim = область по x
    '''
    
    plt.figure(figsize=(15, 7))
    for i in range(sample_count):
        plt.plot(np.arange(sample_size) + 1, estimation[i], color='blue', alpha=0.05)
    if type(left) and type(right) is np.ndarray:
        
        plt.plot(np.arange(sample_size) + 1, left, color='black')
        plt.plot(np.arange(sample_size) + 1, right, color='black')
        labels = [r'$S^2$', r'$\sigma^2$',
                  'Границы доверительного интервала']
        handels = [Line2D([0], [0], color='blue', lw=2),
                   Line2D([0], [0], color='red', lw=2),
                   Line2D([0], [0], color='black', lw=2)]
        
    else: 
        labels = [r'$S^2$', r'$\sigma^2$']
        handels = [Line2D([0], [0], color='blue', lw=2),
                   Line2D([0], [0], color='red', lw=3),]
    
    plt.hlines(np.var(samples2), 0 ,sample_size, color='red')
    plt.title('Поведение оценки при росте размера выбоки')
    plt.xlabel('Размер выборки')
    plt.ylim((0, 5))
    plt.legend(handels, labels)
    plt.xlim((0, xlim));
```

```python
est_plot(estimation, sample_count2, sample_size2)
```

```python
#Границы доверительных интервалов для проверки асимптотической нормальности
right = np.var(samples2) + 2*np.var(samples2)/np.sqrt(np.arange(sample_size2)+1)
left = np.var(samples2) - 2*np.var(samples2)/np.sqrt(np.arange(sample_size2)+1)
```

```python
est_plot(estimation, sample_count2, sample_size2, left=left, right=right)
```
