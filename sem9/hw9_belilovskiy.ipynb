### Задача 1
Пусть $X = ( X_1, ... X_n)$ — выборка из неизвестного распределения $\mathsf{P} \in \{ \mathsf{P}_{\theta} | \theta \in \Theta \}$
и для проверки гипотез $\mathsf{H}_0 : \theta \in [−1, 1]$   vs.  $\mathsf{H}_1 : \theta \notin [−1, 1]$ используется критерий
$S = \{x \in \mathbb{R}^n | T(x) > 1\}$. Предположим, $T(x)$ имеет распределение $\mathcal{N}(\theta, 1)$. Какие
ответы возможны в результате проверки данных гипотез и в каких случаях? Найдите
вероятность ошибки первого рода.

*Указание:*
- Не забудьте, что в определении вероятности ошибки I рода стоит супремум по всем параметрам из основной гипотезы
- Можно показать, где достигается супремум графически. Например, с помощью `sps.norm`, где параметр `loc` отвечает за сдвиг.



Видно, что в данной ситуации возможно у варианта ответов:
$T(x) > 1$ или $T(x) \leq 1$

В первом случае критерий выполнен, а значит мы отвергаем нулевую гипотезу $H_0$. Во втором случае мы не можем отвергнуть нулевую гипотезу по критерию оценки $\Longrightarrow$ получаем *статистически незначимый* результат.

Далее найдём вероятность ошибки первого рода.
$$
\mathbb P(I_S) = \underset{\theta \in \Theta_0}{\sup} \mathbb P(T(X)>1) = 0.5,
$$

Нетрудно понять, что супремум по всем $\theta \in \Theta_0$ будет достигаться при $\theta=1$. Тогда $\mathbb P [T(X)>1]\ =\ 0.5$
Это видно потому что $\theta = \ \longrightarrow$ $\frac 12 -$квантиль распределения $\mathcal{N}(\theta, 1)$

Теперь можно построить график распределения статистики $T(X)$ при различных значениях $\theta\ \in[-1; 1]$

```python
thetas=np.linspace(-1,1,50)
P_thetas=1-sps.norm(loc=thetas).cdf(1) #По определению функции распределения
plt.figure(figsize=(10,7))
plt.plot(thetas, P_thetas, label='Вероятность $T(x)>1$', lw=2)
plt.axvline(1, ymin=0, ymax=0.95, color='black', linestyle='--')
plt.axhline(0.5, xmin=0, xmax=0.95, color='black', linestyle='--', label='Максимальное значение')
plt.legend(fontsize=14)
plt.xlabel(r'$\theta$', fontsize=12)
plt.ylabel(r'$P [T(x)>1]$')
```

Как и ожидалось теоретически, полученный график функции распределения строго возрастает и достигает мактсмального на отрезке $\theta\ \in[-1;1]$ значения в точке $\theta=1$, где $\mathbb P [T(x)>1]\ =\ 0.5$

Необходимо проверить, является ли это распределение с параметром $\lambda=2$. Используя Критерий Вальда, сделайте вывод по данному предположению.

$X_1, ... X_n$ &mdash; выборка из распределения $Exp(\lambda)$.  
Проверьте гипотезу $\mathsf{H}_0\colon \lambda = 2$ vs. $\mathsf{H}_1\colon \lambda \neq 2$  
  

 
 

**1.** Из лекции вы узнали про критерий Вальда.
Для случая двусторонней альтернативы $\mathsf{H}_1\colon \theta \neq \theta_0$ критерий имел следующий вид:
$$\large{S = \left\{ \left|\sqrt{n} \frac{\hat{\theta} - \theta_0}{\hat{\sigma}} \right| > z_{1 - \frac{\alpha}{2}} \right\}}$$

где $\hat{\theta}$ &mdash; асимптотически нормальная оценка $\theta$ с асимптотической дисперсией $\sigma^2(\theta)$, 
$\hat{\sigma}$ &mdash; состоятельная оценка $\sigma(\theta)$.  

Эквивалентный асимптотичсекий доверительный интервал для параметра $\theta$ уровня доверия $1-\alpha$
$$C = \left( \hat{\theta} - \frac{z_{1-\alpha/2} \hat{\sigma}}{\sqrt{n}}, \hat{\theta} + \frac{z_{1-\alpha/2} \hat{\sigma}}{\sqrt{n}}\right)$$

На первой лекции вы получали, что $\frac{1}{\overline{X}}$ &mdash; АНО для параметра $\theta$ c асимптотической дисперсией $\theta^2$


 **Выпишите состоятельную оценку дисперсии и статистику критерия Вальда**
 
 

**Ответ:** Как уже было получено на занятиях ранее, $\widehat{\sigma^2}=\frac{1}{\overline{X}^2}$ - состоятельная оценка дисперсии для выборки из экспоненциального распределения.

 Статистика критерия Вальда: $W(x)=\sqrt{n} \frac{\hat{\theta} - \theta_0}{\hat{\sigma}}$

На лекции вы узнали про p-value &mdash; это вероятность получить при справедливости $H_0$ такое значение статистики $t = T(x)$ или еще более экстремальное, то есть в случае двустороннего критерия
$$p(x) = \mathsf{P}_0(T(X) \geq|t|) + \mathsf{P}_0(T(X) \leq -|t|)$$
Для расчета можно использовать функции из библиотеки `scipy.stats`.

**2**. Оформите функцию подсчета статистики критерия, p-value и доверительного интервала 

```python
def wald_test(sample, theta, estimation_theta, estimation_sigma, alternative='two_sided'):
    """
    param sample: реализация выборки
    param theta: истинное значение параметра
    param estimation_theta: оценка параметра
    param estimation_sigma: оценка асимптотической дисперсии
        оценки estimation_sigma
    param alternative: вид альтернативной гипотезы,
        может принимать одно из значений 'two_sided', 'less', 'greater'

    return statistic
    return p_value
    return conf_int - доверительный интервал
    """

    N = len(sample) #Объём выборки
    Wald = (np.sqrt(N) * (estimation_theta - theta) / estimation_sigma
    )  # статистика критерия Вальда
    statistic = abs(Wald)
    if alternative == 'two_sided':
        alpha = 0.975
        
        z = sps.norm.ppf(alpha)
        interval = (estimation_theta - z / np.sqrt(N) * estimation_sigma,
                    estimation_theta + z / np.sqrt(N) * estimation_sigma)
        p_val = sps.norm.sf(statistic) * 2    
    elif alternative == 'less':
        alpha = 0.95
        z = sps.norm.ppf(alpha)
        interval = (-np.inf, estimation_theta + z / np.sqrt(N) * estimation_sigma)
        p_val = sps.norm.cdf(statistic)   
    elif alternative == 'greater':
        alpha = 0.95
        z = sps.norm.ppf(alpha)
        interval = (estimation_theta - z / np.sqrt(N) * estimation_sigma, np.inf)
        p_val = sps.norm.sf(statistic) 

    print(f'p_value -> {p_val}')
    print(f'Interval -> {interval}')
    print(f'Wald Statistic -> {statistic}') 

    
```

**3**. Проверьте гипотезу с двусторонней альтернативой.

```python
theta = 2
estimation_theta = 1/np.mean(sample)
estimation_sigma = estimation_theta

wald_test(sample, theta, estimation_theta, estimation_sigma, alternative='two_sided')
```

**Вывод:** В данном случае значение `p_value` оказалось меньше чем $\alpha = 0.05$, что позволяет нам опровергнуть нулевую гипотезу $H_0:\ \lambda=2$, так как полученный интервал не содержит 2

**4**. Проверьте гипотезу с правосторонней альтернативой.

```python
wald_test(sample, theta, estimation_theta, estimation_sigma, alternative='greater')
```

**Вывод:** Полученное значение `p_value` позволяет нам отвернуть гипотезую $H_0$. Опять же, нетрудно заметить, что полученный АДИ критерия Вальда не содержит $\lambda = 2$, при этом значение p_value оказалось ниже чем в предыдущем случае, что позволяет нам отвергнуть гипотезу более *уверенно* в данном случае.

**5**. Проверьте гипотезу с левосторонней альтернативой.

```python
wald_test(sample, theta, estimation_theta, estimation_sigma, alternative='less')
```

**Вывод:** В этом случае значение `p_value` (минимальный уровень, при котором $H_0$ может быть отвергнута) оказался значительно больше уровня значимости критерия $\alpha=0.05$, поэтому гипотеза не может быть отвергнута

### Задача 3

На практике часто рассматривают асимптотические критерии, например, критерий Вальда. Такие критерии контролируют вероятность ошибки I рода на уровне $\alpha$ только в пределе, однако для небольших выборок она может сильно отличаться от $\alpha$, причем нередко в большую сторону. Поэтому для таких критериев возникает необходимость в вычислении значения вероятности ошибки I рода или *реального уровня значимости* при конечной выборке.

Зачастую посчитать точное значение реального уровня значимости довольно сложно, поэтому вместо него вычисляют оценку с помощью сэмплирования по методу Монте-Карло.

Пусть гипотеза $\mathsf{H}_0$ простая, то есть $\mathsf{H}_0\colon \mathsf{P} = \mathsf{P}_0$. Для ее проверки по конечной выборке $X_1, ..., X_n$ используется критерий $S$. **Опишите схему оценки реального уровня** значимости критерия $S$ методом Монте-Карло.


**Ответ:** Можно сгенерировать $n$ выборок из $\mathscr{P}_0$ и посчитать для них статистику критерия. Далее считаем количество статистик, удовлетвояющих критерию $S$ то есть отвергающих $H_0$, например $S = \{x \in \mathbb{R}^n | T(x) > 1\}$. Посчитаем долю всех статистик, удовлетворяющих $S$, разделив полученное число на $n$. Полученный результат и будет реальным уровнем значимости критерия $S$ с учётом погрешности $\delta = \frac{1}{\sqrt{n}}$

Если гипотеза $\mathsf{H}_0$ сложная, то есть $\mathsf{H}_0\colon \mathsf{P} \in \mathscr{P}_0$, то можно повторить такую процедуру несколько раз для разных $\mathsf{P} \in \mathscr{P}_0$.

Рассмотрим альтернативную гипотезу $\mathsf{H}_1\colon \mathsf{P} \in \mathscr{P}_1$. **Опишите схему оценки мощности** критерия $S$ методом Монте-Карло.

**Ответ:** В принципе мы поступаем аналогично прошлому пункту. Но теперь нужно сгенерировать $n$ выборок из распределения $\mathscr{P}_1$ и посчитать долю статистик, не удовлетворяющих критерию $S$, отвергающему данную гипотезу

Пусть $X_1, ..., X_n$ &mdash; выборка из пуассоновского распределения с параметром $\theta$. Рассмотрим гипотезы $\mathsf{H}_0\colon \theta = 1\ \ vs.\ \ \mathsf{H}_1\colon \theta \not= 1$ и критерий Вальда для их проверки.

**1.** **Оцените реальный уровень значимости** критерия для $ n = 50$.

Сгенерируем $10^5$ выборок из Распределения Пуассона размера n

```python
sample_count=100000
n=50
samples=sps.poisson(mu=1).rvs(size=(sample_count,n))
```

Оценка параметра $\theta$: $\widehat{\theta}\ =\ \overline{X}$

```python
theta_est=np.mean(samples, axis=1)
theta = 1 #Гипотеза
```

Воспользуемся тем же способом, что и в прошлом задании, когда мы считали долю попаданий $\theta$ в доверительный интервал, что эквивалентно $\mathbb P [T(x)>1]$

```python
alpha=1-0.05/2
z=sps.norm.ppf(alpha)
left = theta_est - z / np.sqrt(n) * np.sqrt(theta_est) #Левая граница
right = theta_est + z / np.sqrt(n) * np.sqrt(theta_est) #Правая граница

counts = ((theta>=left) & (theta <= right)).sum() #Количество попаданий
un_counts=sample_count - counts #Количество НЕпопаданий
print(f' Реальный уровень значимости: {un_counts/sample_count} \n',
      f'Количество выборок: {sample_count} \n',
      f'Число НЕпопаданий в интервал: {un_counts} \n',
      f'Размер выборки: {n}')
```

**Вывод:** Размера выборки в 50 элементов уже достаточно для того, чтобы с высокой точностью оценивать реальный уровень значимости асимптотическим

**Решение:**

**Пункт 1**

Рассмотрим указанную статстику: $$T(X) = \sqrt{n - 1} \frac{\overline{X}}{s} = \sqrt{n} \frac{\overline{X}}{\sigma} \frac{\sqrt{(n-1)\sigma^2}}{\sqrt{nS^2}} = \sqrt{n} \frac{\overline{X}}{\sigma} \frac{1}{\sqrt{\frac{S_n^2}{\sigma^2}}}$$
Где $S_n^2 = \frac{1}{n-1} \sum\limits_{i=1}^n (x_i - \overline{x})^2$ - несмещённая оценка дисперсии $\mathcal{N}(a, \sigma^2)$

Множитель $\sqrt{n} \frac{\overline{X}}{\sigma}$ имеет распределние $\mathcal{N}(0, 1)$, а величина $\frac{S_n^2}{\sigma^2}$ - Распределение хи-квадрат $\chi^2(n-1)$
